```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %pip install xlrd

df = pd.read_excel("RNA__RNA_seq_composite_expression.xls", skiprows=10, header=0)
df = df.set_index("Gene name d")
```

```python
df
```

```python
df.info()
```

```python
dft = np.transpose(df)
dfnew  = dft.iloc[5:,:]
dfnew
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
pca.fit_transform(dfnew)
pca_score = pca.explained_variance_ratio_

pca = PCA(n_components=2)
grp2 = pca.fit_transform(dfnew)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(np.arange(1,21),pca_score,marker="o")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title(f"Scree Plot, variance explained by PC1 & PC2 : {pca_score[0]+pca_score[1]}")
plt.subplot(1,2,2)
plt.scatter(grp2[:,0],grp2[:,1])
plt.xlabel("PC 1")
plt.ylabel("PC 2")
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, homogeneity_completeness_v_measure
wcss = []
kmeans_score = []
for i in range (2,31) :
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=11)
    kmeans.fit(grp2)
    wcss.append(kmeans.inertia_)
    kmeans_labels = kmeans.predict(grp2)
    kmeans_score.append(silhouette_score(grp2, kmeans_labels))
```

```python
from sklearn.cluster import AgglomerativeClustering
aglo_score = []
for i in range (2,31) :
    aglo = AgglomerativeClustering(n_clusters = i)
    aglo_labels = aglo.fit_predict(grp2)
    aglo_score.append(silhouette_score(grp2, aglo_labels))
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.plot(np.arange(2,31),wcss,marker="o")
plt.ylabel("WCSS for k-means")
plt.xlabel("No. of clusters")
plt.subplot(1,3,2)
plt.bar(np.arange(2,31),kmeans_score)
plt.ylabel("Silhouette score for k-means")
plt.xlabel("No. of clusters")
plt.subplot(1,3,3)
plt.bar(np.arange(2,31),aglo_score)
plt.ylabel("Silhouette score for agglomerative")
plt.xlabel("No. of clusters")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
grp2 = pca.fit_transform(dfnew)

kmeans = KMeans(n_clusters=7, init="k-means++")
kmeans_labels2 = kmeans.fit_predict(grp2)
c = kmeans.cluster_centers_

aglo = AgglomerativeClustering(n_clusters = 3)
aglo_labels = aglo.fit_predict(grp2)
```

```python
cmap = plt.get_cmap("tab20", 7)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.scatter(grp2[:,0],grp2[:,1],c=kmeans_labels2,cmap=cmap,alpha=0.6,s=120,label="cluster elements")
plt.scatter(c[:,0],c[:,1],c=range(7),marker="*",cmap=cmap,s=90,edgecolor="black",label="cluster centroids")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("k-means clusters")
plt.legend(loc="lower right")
plt.subplot(1,2,2)
plt.scatter(grp2[:,0], grp2[:,1], c=aglo_labels, cmap="tab20", alpha=0.6,s=120)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("agglomerative clusters")
```

```python
targetdf = pd.read_excel("NCI60_CELL_LINE_METADATA.xls", skiprows=7, header=0)
targetdf = targetdf.set_index("Cell Line Name")
targetdf
```

```python
target = np.array(targetdf.iloc[:60,0])
target
```

```python
from sklearn.metrics import adjusted_rand_score
ars = []
ars1 = []
ars2 = []
for i in range (1,31) :
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=11)
    kmeans.fit(dfnew)
    kmeans_labels3 = kmeans.predict(dfnew)
    ars.append(adjusted_rand_score(target,kmeans_labels3))
    kmeans1 = KMeans(n_clusters=i, init="k-means++", random_state=11)
    kmeans1.fit(grp2)
    kmeans_labels4 = kmeans1.predict(grp2)
    ars1.append(adjusted_rand_score(target,kmeans_labels4))
    aglo1 =  AgglomerativeClustering(n_clusters=i)
    aglo_label2 = aglo1.fit_predict(dfnew)
    ars2.append(adjusted_rand_score(target,aglo_label2))
```

```python
plt.plot(np.arange(1,31),ars1,label="k-means with PC-1 & PC-2") 
plt.plot(np.arange(1,31),ars,label="k-means without PCA")
plt.plot(np.arange(1,31),ars2,label="agglomerative without PCA")
plt.ylabel("Adjusted Rand Score")
plt.xlabel("No. of clusters")
plt.legend(loc="upper left")
plt.show()
```

```python
aglo2 =  AgglomerativeClustering(n_clusters=len(np.unique(target)))
aglo_label3 = aglo2.fit_predict(dfnew)
hcv = homogeneity_completeness_v_measure(target,aglo_label3)
print(f"Homogeneity Score : {hcv[0]}")
print(f"Completeness Score : {hcv[1]}")
print(f"V-measure Score : {hcv[2]}")
```
